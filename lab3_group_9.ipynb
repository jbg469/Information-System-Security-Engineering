## Part a - Plotting Sales Data

````python
import matplotlib.pyplot as plt
import csv

header = []
data = []
months = []
sales = [[], [], [], [], [], []]
with open('product_sales_Data.csv') as f:
    reader = csv.reader(f, delimiter = ';')
    header = next(reader)
    data = [row for row in reader]
    for row in data:
        months.append(row[0])
        sales[0].append(int(row[1]))
        sales[1].append(int(row[2]))
        sales[2].append(int(row[3]))
        sales[3].append(int(row[4]))
        sales[4].append(int(row[5]))
        sales[5].append(int(row[6]))

plt.figure(figsize=(12.8,9.6)) 
for i in range(6):
    plt.plot(months, sales[i], linestyle = 'dashed', marker = '*', label=header[i+1])
plt.xlabel('Month', fontsize = 15)
plt.ylabel('Sales', fontsize = 15)
plt.title('Sales of products', fontsize = 20)
plt.legend(loc=2)
plt.show()
````

## Part e - What is multithreading?

Multithreading is a way to execute programs that allows a single process to have multiple threads of execution in other words different instructions can run at once on different threads. Threads are individual processes that can be executed independently. It's a programs way of "multitasking".
https://totalview.io/blog/multithreading-multithreaded-applications

````python
import threading
from math import sqrt
 
def factorial(n) -> int:
    # function to print factorial of given num
    factorial = 1
    # check if the number is negative, positive or zero
    if n < 0:
      return 0;
    elif n == 0:
      return 1
    else:
      for i in range(1, n + 1):
          factorial = factorial*i
    print(factorial)
    return factorial
 
def root(n) -> int:
    # function to print root of given num
    print(sqrt(n))
    return sqrt(n)
 
if __name__ =="__main__":
    # creating thread
    t1 = threading.Thread(target= factorial, args=(21,))
    print(t1)
    t2 = threading.Thread(target= root, args=(91,))
    # starting thread 1
    t1.start()
    # starting thread 2
    t2.start()
    # wait until thread 1 is completely executed
    t1.join()
    # wait until thread 2 is completely executed
    t2.join()
    # both threads completely executed

    print("This was multithreaded "+ str(t1) + " " + str(t2) + " .")

    #threading example retrieved from https://www.geeksforgeeks.org/multithreading-python-set-1/
````

## Part a - Differences between symmetric and asymmetric encryption

Some advantages of using public key cryptography is that its fast as there is only one key used and the shorter key lengths make managemetn of keys easier.
A downfall of using symmetric key cryptography is that there is more of a risk of being intercepted by a MITM as the key must be shared with anyone looking to decrypt.

Asymmetric key cryptography is better overall as it is more secure as there is a public key and private key that are used to encrypt and decrypt. As long as the private keys between sender and recipients aren't compromised, no third party will be able to decrypt messages sent with asymmetric encryption. Some cons of using asymmetric key cryptography are longer key lengths and that its slower lower for sharing messages. A session key produced with asymmetric cryptography solves some authentication issues that come with symetric cryptography.

## Part b - Explanation on what each class of code does and how Pydantic is used

Pydantic maps data into classes so that no value is accidentally lost for example in dictionaries where a value could be part of a key. Pydantic gives user friendly errors when data is invalid and provides type hints at run time. Based on the documentation it is a fast way to parse data structures and validate data structures. It has been used by Microsoft, Uber, the NSA and more. 

````python
"""
Basic example showing how to read and validate data from a file using Pydantic.
"""

"""
Sample data file:
[
  {
    "title": "Zero to One",
    "subtitle": "Notes on Startups, or How to Build the Future",
    "author": "Peter Thiel",
    "publisher": "Ballantine Books",
    "isbn_10": "0753555190",
    "isbn_13": "978-0753555194",
    "price": 14.29,
    "author2": {
      "name": "Peter Thiel",
      "verified": true
    }
  }
]
"""

import json
from typing import List, Optional

import pydantic
class ISBNMissingError(Exception):
    def __init__(self, title: str, message: str) -> None:
        self.title = title
        self.message = message
        super().__init__(message)

"""
This class initalizes an instance of an exception message should there be no ISBM in the data set 
"""
class ISBN10FormatError(Exception):

    def __init__(self, value: str, message: str) -> None:
        self.value = value
        self.message = message
        super().__init__(message)
"""
This class initalizes an instance of an exception message should there be an error in the ISBM format.
"""

class Author(pydantic.BaseModel):
    name: str
    verified: bool
"""
This class constructs an Author object with a string for the name and a boolean to check wether they are verfied or not
"""

class Book(pydantic.BaseModel):
    """
    This class maps the desired fields from the data model into a class based on needed specifications.
    It does the same for the Author 2 field when there is one given the Optional tag implies that that book may or may not have that property it could be missing. 
    """
    title: str
    author: str
    publisher: str
    price: float
    isbn_10: Optional[str]
    isbn_13: Optional[str]
    subtitle: Optional[str]
    author2: Optional[Author]

    @pydantic.root_validator(pre=True)
    @classmethod
    def check_isbn_10_or_13(cls, values):
        if "isbn_10" not in values and "isbn_13" not in values:
            raise ISBNMissingError(
                title=values["title"],
                message="Document should have either an ISBN10 or ISBN13",
            )
        return values

    @pydantic.validator("isbn_10")
    @classmethod
    def isbn_10_valid(cls, value) -> None:
        chars = [c for c in value if c in "0123456789Xx"]
        if len(chars) != 10:
            raise ISBN10FormatError(value=value, message="ISBN10 should be 10 digits.")

        def char_to_int(char: str) -> int:
            if char in "Xx":
                return 10
            return int(char)

        if sum((10 - i) * char_to_int(x) for i, x in enumerate(chars)) % 11 != 0:
            raise ISBN10FormatError(
                value=value, message="ISBN10 digit sum should be divisible by 11."
            )
        return value

    class Config:

        allow_mutation = False
        anystr_lower = True


def main() -> None:

    # Read data from a JSON file
    with open("./data.json") as file:
        data = json.load(file)
        books: List[Book] = [Book(**item) for item in data]
        print(books)
        print(books[0])
        print(books[0].dict(exclude={"price"}))
        print(books[1].copy())
"""
Here in the function we sort through the data in data.json and pydantic will recognize each instance of a book as a pydantic.
Basemodel and automagicaally do the mapping 
"""

if __name__ == "__main__":
    main()
````
