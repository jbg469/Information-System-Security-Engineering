```python
# The following piece of code is swapping the first two words with the last two words
string="The quick brown fox jumps over the lazy dog"
arr = string.split(" ") # this is converting the sentence into an array of words by splitting it based on spaces
temp=[]
temp.append(arr[0]) # adding "The" into the array at index 0
temp.append(arr[1]) # adding "quick" into the array at index 1
arr[0]= arr[len(arr)-2] # replacing "The" with "lazy"
arr[1]= arr[len(arr)-1] # replacing "quick" with "dog"
arr[len(arr)-2] = temp[0] # replacing "lazy" with "The"
arr[len(arr)-1] = temp[1] # replacing "dog" with "quick"
string=' '.join(arr) # remaking the sentence by joining the words with a space as the seperator between the words
```

```python
# This function takes in a number and returns the reversed version of the number
def dummy(var):
    x = str(var)
    y = len(x)-1
    z = ''
    while y >= 0:
        z = z + x[y]
        y -=1
    return int(z)
```

```python
array = [121, 123, 414, 866, 988, 101]
for a in array:
    b = dummy(a)
    if b == a:
        print(a,"is a palindrome")
    else:
        print(a,"is not a palindrome")
```

```python
def Calculate_Discount(price):
    if price >= 500:
        final_price = price * (70/100)
    elif price >= 350 and price < 500:
        final_price = price * (80/100)
    elif price >= 150 and price < 350:
        final_price = price * (90/100)
    elif price >= 10 and price < 150:
        final_price = price * (95/100)
    elif price > 0 and price < 10:
        final_price = price * (50/100)
    else:
        final_price = "Price cannot be negative. Please enter a positive value." 
    return final_price

array = [440, 265, 144, -10, 8]
for a in array:
    new_price = Calculate_Discount(a)
    if isinstance(new_price, str):
        print(a,"-->",new_price)
    else:
        print("The new price for",a,"is",new_price)
```
